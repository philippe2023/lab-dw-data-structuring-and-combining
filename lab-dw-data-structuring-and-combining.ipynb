```python
# Your code goes here
# inport the data and create 3 datasets
import pandas as pd
insurance_data_1_url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
insurance_data_2_url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
insurance_data_3_url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'

df_1 = pd.read_csv(insurance_data_1_url)
df_2 = pd.read_csv(insurance_data_2_url)
df_3 = pd.read_csv(insurance_data_3_url)
```

```python
# initial analysis of the 3 df
display(df_1.head(2), df_2.head(2), df_3.head(2),df_1.info(), df_2.info(), df_3.info())
```

```python
complete_df = pd.concat([df_1, df_2, df_3], axis=0)
display(complete_df)
```

```python
# clean the complete df
# Function to rename columns to lowercase
def rename_col(complete_df):
    complete_df = complete_df.rename(columns={column: column.lower() for column in complete_df.columns})
    return complete_df
# Function to replace white spaces with underscores and convert to lowercase
def lower_snake_col(complete_df):
    complete_df = complete_df.rename(columns={column: column.replace(" ", "_").lower() for column in complete_df.columns})
    return complete_df
# Function to replace "State" with "st"
def new_state_name(complete_df):
    complete_df = complete_df.rename(columns={column: column.replace("state", "st").lower() for column in complete_df.columns})
    return complete_df
# Function to apply all transformations
def new_df(complete_df):
    complete_df = rename_col(complete_df)  # Convert column names to lowercase
    complete_df = lower_snake_col(complete_df)  # Replace spaces with underscores
    complete_df = new_state_name(complete_df)  # Replace "State" with "st"
    return complete_df
```

```python
updated_df = new_df(complete_df)
display(updated_df)
```

```python
# correct gender column
def new_gender(updated_df):
    gender_mapping = {'F':'F', 'Female':'F', 'f':'F', 'female':'F', 'FEMALE':'F',
                          'M':'M', 'Male':'M', 'm':'M', 'male':'M', 'MALE':'M'}
    updated_df['gender'] = updated_df['gender'].replace(gender_mapping)
    return updated_df

# state abbreviations
def new_state(updated_df):
    state_mapping = {"AZ": "Arizona", "Cali": "California", "WA": "Washington"}
    updated_df['st'] = updated_df['st'].replace(state_mapping)
    return updated_df

# education, "Bachelors" to "Bachelor"
def new_education(updated_df):
    education_mapping = {"Bachelors": "Bachelor"}
    updated_df['education'] = updated_df['education'].replace(education_mapping)
    return updated_df

# Customer Lifetime Value, delete the % character
def new_customer_lifetime_value(updated_df):
    updated_df['customer_lifetime_value'] = updated_df['customer_lifetime_value'].str.replace('%', '')
    updated_df['customer_lifetime_value'] = pd.to_numeric(updated_df['customer_lifetime_value'], errors='coerce')
    return updated_df

# Customer Lifetime Value, delete the % character
def new_vehicle_class(updated_df):
    updated_df['customer_lifetime_value'] = updated_df['customer_lifetime_value'].str.replace('%', '')
    updated_df['customer_lifetime_value'] = pd.to_numeric(updated_df['customer_lifetime_value'], errors='coerce')
    return updated_df

# In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"
def simplify_vehicle(updated_df):
    vehicle_mapping = {'Sports Car': 'Luxury', 'Luxury SUV': 'Luxury', 'Luxury Car': 'Luxury'}
    updated_df['vehicle_class'] = updated_df['vehicle_class'].replace(vehicle_mapping)
    return updated_df

def clean_df(updated_df):
    df2 = updated_df.copy()
    df2 = new_gender(df2)
    df2 = new_state(df2)
    df2 = new_education(df2)
    df2 = new_customer_lifetime_value(df2)
    df2 = simplify_vehicle(df2)

    return df2
```

```python
cleaned_df = clean_df(updated_df)
display(cleaned_df)
```

```python
def final_df(cleaned_df):
    dropped_nan_df = cleaned_df.dropna(how='all').dropna(how='all')
    return dropped_nan_df
```

```python
final_table = final_df(cleaned_df)
display(final_table)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# Create a pivot table to calculate total revenue by sales channel
sales_channel_pivot = pd.pivot_table(c2_df, values='total_claim_amount', index='sales_channel', aggfunc='sum').round(2)

# Display the resulting pivot table
print(sales_channel_pivot)
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Create a pivot table for average customer lifetime value by gender and education level
clv_pivot = pd.pivot_table(c2_df, values='customer_lifetime_value', index='gender', columns='education', aggfunc='mean').round(2)

# Display the resulting pivot table
print(clv_pivot)
```

Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Your code goes here
# Group by policy type and month, and count the number of complaints
complaints_summary = c2_df.groupby(['policy_type', 'month'])['number_of_open_complaints'].sum().reset_index()

# Display the summary table
print(complaints_summary)
```
